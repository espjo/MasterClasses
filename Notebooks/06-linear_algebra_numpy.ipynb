## A Linear Regression example with numpy

Now, we are ready to implement our own linear regression example. 

In linear regression, our hypothesis function $h_\theta$ is:

$$h_\theta(x) = \theta_0 + \theta_1x$$

And, as we are doing regression, our cost function is: 

$$J(\theta_0,\theta_1) = \frac{1}{m}\sum_{i=1}^m(\hat{y}_i-y_i)^2 = \frac{1}{m}\sum_{i=1}^m(h_\theta(x_i)-y_i)^2 $$

### Generate dummy data

```python
thr0 = 50
thr1 = 1.3

xs = np.random.normal(30.0, 10.0, 250)
ys = thr0 + xs * thr1 + np.random.normal(0, 5.0, 250)
plt.scatter(xs, ys)
```

That is the spread that we will try to aproximate with our line.

### Write the cost function

```python
def cost_function(th0, th1):
    return (((th0 + th1 * xs) - ys) ** 2).sum()

th0_hypothesis = 0
th1_hypothesis = 0
print(f"Example of cost fuction value for theta0 = {th0_hypothesis} and theta1 = {th1_hypothesis}, J(th0, th1) = {cost_function(th0_hypothesis, th1_hypothesis)}")
```

### Gradient descent

Remember, we have to descend in the direction of the steepest gradient. For that, we need to now what direction the gradient points!

### Partial differentials of the cost function

$$\frac{\partial}{\partial\theta_0} = \frac{1}{m}\sum_{i=1}^m(h_\theta(x_i)-y_i)$$

$$\frac{\partial}{\partial\theta_1} = \frac{1}{m}\sum_{i=1}^m(h_\theta(x_i)-y_i) \cdot x_i$$

```python
def derivative_cost_function_in_theta0(th0, th1):
    return (((th0 + th1 * xs) - ys).sum()) / xs.shape[0]

def derivative_cost_function_in_theta1(th0, th1):
    return ((((th0 + th1 * xs) - ys) * xs).sum()) / xs.shape[0]
```

```python
samples = 100
th0_hypothesis = 0
th1_hypothesis = 1

dcf_theta0 = np.vectorize(derivative_cost_function_in_theta0)

th0s = np.array([th0_hypothesis] * samples)
th1s = np.linspace(-5, 5, num = samples)
plt.plot(th1s, dcf_theta0(th0s, th1s), label = f"theta0 = {th0_hypothesis}, theta1 = [{th1s[0]}, {th1s[-1]}]", color = "b")

th0s = np.linspace(-5, 5, num = samples)
th1s = np.array([th1_hypothesis] * samples)
plt.plot(th0s, dcf_theta0(th0s, th1s), label = f"theta0 = [{th0s[0]}, {th0s[-1]}], theta1 = {th1_hypothesis}", color="r")
plt.title("Gradient in theta0")
plt.legend()
```

```python
dcf_theta1 = np.vectorize(derivative_cost_function_in_theta1)

th0s = np.array([th0_hypothesis] * samples)
th1s = np.linspace(-5, 5, num = samples)
plt.plot(th1s, dcf_theta1(th0s, th1s), label = f"theta0 = {th0_hypothesis}, theta1 = [{th1s[0]}, {th1s[-1]}]", color = "b")

th0s = np.linspace(-5, 5, num = samples)
th1s = np.array([th1_hypothesis] * samples)
plt.plot(th0s, dcf_theta1(th0s, th1s), label = f"theta0 = [{th0s[0]}, {th0s[-1]}], theta1 = {th1_hypothesis}", color="r")
plt.title("Gradient in theta1")
plt.legend()
```

### Implementing gradient descent

Now we are ready to implement the actual gradient descent. Remember, the algorithm is:

- Initialize variables

- Compute cost function
iterations
- Compute gradients

- Update variables: gradient times learning rate (alpha)

- Repeat until convergence: cost at iteration n-1 $\sim$ cost at iteration n

```python
import time

# Initialize variables
th0_hypothesis = 0
th1_hypothesis = 0
learning_rate = 0.001
acceptable_convergence = 0.0001

# Gradients
gradient_in_theta0 = derivative_cost_function_in_theta0
gradient_in_theta1 = derivative_cost_function_in_theta1

def trace(function):
    def wrapper(*args):
        start_time = time.perf_counter()
        value = function(*args)
        end_time = time.perf_counter()
        run_time = end_time - start_time
        print("Finished {}() in {:.4f} secs\n".format(function.__name__, run_time))
        return value
    return wrapper
    
@trace
def convergence(th0_hypothesis = 0, th1_hypothesis = 0, acceptable_convergence = 0.0001):
    _th0 = th0_hypothesis
    _th1 = th1_hypothesis
    iteration = 1
    while True:
        th0 = _th0 - learning_rate * gradient_in_theta0(_th0, _th1)
        th1 = _th1 - learning_rate * gradient_in_theta1(_th0, _th1)
        convergence = abs(cost_function(th0, th1) - cost_function(_th0, _th1))
        if convergence < acceptable_convergence:
            break
        _th0 = th0
        _th1 = th1
        iteration += 1
    print(f"Acceptable convergence = {acceptable_convergence}")
    print(f"Cost function convergence = {convergence}")
    print(f"\tCost function value = {cost_function(th0, th1)}")
    print(f"\tth0 = {th0}") 
    print(f"\tth1 = {th1}")
    print(f"\tIterations = {iteration}")
    return th0, th1
```

```python
th0, th1 = convergence(th0_hypothesis, th1_hypothesis, 0.0001)
th0, th1 = convergence(th0_hypothesis, th1_hypothesis, 0.00001)
th0, th1 = convergence(th0_hypothesis, th1_hypothesis, 0.000001)
th0, th1 = convergence(th0_hypothesis, th1_hypothesis, 0.0000001)
```
